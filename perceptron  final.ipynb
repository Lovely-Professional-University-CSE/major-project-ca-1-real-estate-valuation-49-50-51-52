```python
##learning rate is basically used for moving the weights toward the global mimima
learning_rate = np.random.rand(1,1)
learning_rate = learning_rate[0,0]
learning_rate
```

```python
# defining the activation function
def activation_function(x):
    return (1/(1+np.exp(-x)))
```

```python
## the algorithm for perceptron is 
## initialize the weights
## add the weights((Xi*weightsi)+bais))
## pass through an activations function
##check for thereshoud
##give the output

threshold = 0.5
for i in range(len(x)):
    s = 0
    for j in range(5):
        ##summation
        s = s + (x[i,j]*weights[j,0])
    s = s + bais[0,0]
    ##activation function
    ans = activation_function(s)
    d = 0
    ##checkig for threshold
    if ans>threshold:
        d = 1
    else:
        d = 0 
    ##if thery are equal
    if d == y[i]:
        ##key the same weights
        weights = weights
        bais = bais
    else:
        ##update the weights
        r=1
        for k in range(5):
            newweights[k,0] = learning_rate*x[i,k]*y[i]
        newbais =  learning_rate*y[i]
        weights = weights + newweights
        bais = bais + newbais
        
        
        print('weights\n',weights)
        print('bais\n',bais)
## and the final updated weights and bais are shown


print('The final updated weights are\n',weights)
print('The final updated bais is\n',bais)
```
